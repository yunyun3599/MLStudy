# Ensemble & Random Forest

## Ensemble
약한 분류기들을 결합하여 강 분류기로 만드는 것

### Ensemble 종류
1. Bagging
2. Boosting
3. Stacking

## Bagging
= Bootstrap + Aggregation
- Bootsrap: 약한 분류기를 생성하는 방법
- Aggregation: 생성된 분류기를 결합하는 방법

### Bootstrap
Training Data에서 여러번 복원 추출하는 Random Sampling 기법  
추출된 샘플들을 Bootstrap Sample이라고 함  
image.png

Out-Of-Bag Sample: Bootstrap을 통해 Sample들을 뽑아냈을 때 아무 데에도 뽑히지 않은 데이터 (이론적으로 36.8%의 샘플이 out-of-bag 샘플)  

### OOB (Out-Of-Bag) 평가
추출되지 않은 샘플을 이용해 Cross Validation(교차 검증)에서 Validation 데이터로 사용 가능

### 약분류기 생성
추출된 부트스트랩 샘플마다 약분류기를 학습  
image.png

### Aggregation
생성된 약분류기들의 예측 결과를 Voting을 통해 결합  

#### Voting의 종류
1. Hard Voting
  - 예측한 결과값 중 다수의 분류기가 결정한 값을 최종 예측값으로 선정
    image.png
    
2. Soft Voting  
  - 분류기가 예측한 확률 값의 평균으로 결정
    image.png

### Bagging의 장점
분산을 줄이는 효과가 있음  
- 원래 추정 모델이 불안정하면 분산 감소 효과를 얻을 수 있음
- 과대 적합이 심한(High Variance) 모델에 적합

## Random Forest
= Decision Tree + Bagging  
- 분산이 큰 Decision Tree + 분산을 줄일 수 있는 Bagging

### Random Forest와 무작위성(Randomness)
- 무작위성을 더 강조하여 의사결정나무들이 서로 조금씩 다른 특성을 가짐  
  - 변수가 20개가 있다면 5개의 변수만 선택해서 의사결정나무를 생성
- 의사결정나무의 예측들이 비상관화되어 일반화 성능을 향상시킴

### Random Forest 장단점
장점  
- 의사결정나무의 단점인 Overfitting을 해결
- 노이즈 데이터에 영향을 크게 받지 않음
- 의사결정나무 모델보다 복잡도가 적음

단점 
- 모델의 예측 결과를 해석하고 이해하기 어려움
