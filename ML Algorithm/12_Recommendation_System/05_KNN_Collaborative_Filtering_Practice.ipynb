# KNN 협업 필터링 실습  
유저, 영화 평점 데이터를 이용해 유저가 아직 평가하지 않은 영화 추천을 진행  

```python
import pandas as pd
import numpy as np

np.random.seed(2021)
```

## 1. Data
### 1.1 Data Load
데이터에서유저 고유 아이디를 나타내는 `userId`, 영화 고유 아이디를 나타내는 `movieId`, 유저가 영화를 평가한 점수 `rating` 컬럼을 이용

```python
ratings = pd.read_csv("/content/drive/MyDrive/MLStudy/ML Algorithm/12_Recommendation_System/ratings_small.csv")
ratings = ratings[["userId", "movieId", "rating"]]
```

```python
ratings.head()
```

다른 두 데이터를 이용해 `ratings` 데이터의 `movieId`에 맞는 영화 제목을 얻음  

```python
movies = pd.read_csv("/content/drive/MyDrive/MLStudy/ML Algorithm/12_Recommendation_System/movies_metadata.csv")
links = pd.read_csv("/content/drive/MyDrive/MLStudy/ML Algorithm/12_Recommendation_System/links_small.csv")
```

## 1.2 Data Preproccessing
`movies` 데이터에서 "tt숫자"로 이루어진 `imdb_id`에서 숫자 부분과 links 데이터의 "숫자"로 이루어진 `imdbId`와 연결

```python
movies = movies.fillna('')
movies = movies[movies["imdb_id"].str.startswith('tt')]
movies["imdbId"] = movies["imdb_id"].apply(lambda x: int(x[2:]))
movies = movies.merge(links, on="imdbId")
```

```python
movies = movies[["title", "movieId"]]
movies = movies.set_index("movieId")
```

```python
movies.head()
```

`pivot`함수를 이용해 유저 아이디가 인덱스이고, 영화 아이디가 컬럼, 값이 평가 점수인 `user_movie_matrix`를 만듦

```python
user_movie_matrix = ratings.pivot(
    index="userId",
    columns="movieId",
    values="rating"
)
```

```python
user_movie_matrix.iloc[-5:, -5:]
```

유저가 평가하지 않은 영화에 대해서 결측값을 0으로 대체

```python
user_movie_matrix = user_movie_matrix.fillna(0)
```

```python
user_movie_matrix.shape
```

## 2. KNN Basic
k=5인 KNN Basic을 이용해 유저 "124"가 아직 평가하지 않은 영화 "648"에 대한 점수를 예측

```python
k = 5
user_i = 124
movie_id = 648
```

### 2.1 유저 간의 유사도를 계산  
`cosine_similarity`함수를 이용해 유저별 코사인 유사도를 계산

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_movie_matrix)
```

```python
user_similarity.shape
```

```python
user_similarity[:10, :10]
```

```python
user_similarity = pd.DataFrame(
    data=user_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)
```

```python
user_similarity.head(5)
```

### 2.2 아이템 i를 평가한 유저들 중에서 유저 u와 비슷한 유저 k명을 찾음
유저 "124"와 유사한 다른 유저 k명 찾기

```python
user_i_similarity = user_similarity.loc[user_i]
```

```python
user_i_similarity
```

```python
user_i_similarity = user_i_similarity.sort_values(ascending=False)
```

```python
user_i_similarity
```

유사도 상위 k명의 유사도와 id를 추출  
이 때 가장 유사도가 높은 id는 user_i로 제외

```python
top_k_similarity = user_i_similarity[1 : k+1]
top_k_similar_user_ids = top_k_similarity.index
```

```python
top_k_similar_user_ids
```

### 2.3 K명의 유사한 유저들이 아이템 i에 평가한 선호도를 유사도 기준으로 가중 평균

```python
top_k_similar_ratings = user_movie_matrix.loc[top_k_similar_user_ids, movie_id]
```

```python
movie_id
```

```python
top_k_similar_ratings
```

평가 점수가 있는 유저에 대한 weight를 추출

```python
top_k_weight = (top_k_similar_ratings > 0) * top_k_similarity
```

```python
top_k_weight
```

유사도가 곱해진 평가 점수의 합을 유사도 합으로 나눔

```python
top_k_weighted_ratings = top_k_similar_ratings * top_k_weight
weighted_rating = top_k_weighted_ratings.sum()
weight = top_k_weight.sum()
```

```python
weight
```

weight가 0보다 작은 경우는 유저 모두가 평가하지 않은 경우

```python
if weight > 0:
  prediction_rating = weighted_rating / weight
else:
  prediction_rating = 0
```

```python
prediction_rating
```

### 2.4 예측 선호도가 높은 아이템을 유저에게 추천  
모든 영화에 대해서 점수를 예측하고 예측 평가 점수가 높은 영화를 유저에게 추천

#### 2.4.1 선호도 계산

```python
prediction_dict = {}

# 모든 영화 아이디에 대해 평점 예측
for movie_id in user_movie_matrix.columns:

  # 이미 유저가 평가한 경우 제외
  if user_movie_matrix.loc[user_i, movie_id] > 0:
    continue

  top_k_similar_ratings = user_movie_matrix.loc[top_k_similar_user_ids, movie_id]

  top_k_weighted_ratings = top_k_similar_ratings * top_k_similarity
  top_k_weight = (top_k_similar_ratings > 0) * top_k_similarity

  weighted_rating = top_k_weighted_ratings.sum()
  weight = top_k_weight.sum()

  if weight > 0:
    prediction_rating = weighted_rating / weight
  else:
    prediction_rating = 0

  # 영화 아이디별로 예측 평가 점수 저장
  prediction_dict[movie_id] = prediction_rating
```

영화 아이디별 예측 평가 점수를 내림차순으로 정렬

```python
prediction = pd.Series(prediction_dict).sort_values(ascending=False)
```

#### 2.4.2 상위 아이템 추출
예측 평가 점수 상위 10개의 영화 아이디를 추출

```python
recommend = prediction[:10].index
```

```python
movies.loc[recommend]
```

```python
movies.loc[recommend]
```

## 3. KNN with Means
k가 5인 KNN Basic을 이용해 유저 "124"가 아직 평가하지 않은 영화 "31"에 대한 점수를 예측

```python
user_id = 124
k = 5
movie_i = 648
```

pivot 함수를 이용해 영화 아이디가 인덱스이고, 유저 아이디가 컬럼, 값이 평가 점수인 movie_user_matrix를 만듦  
결측값은 0으로 대체

```python
movie_user_matrix = ratings.pivot(
    index="movieId",
    columns="userId",
    values="rating"
)
movie_user_matrix = movie_user_matrix.fillna(0)
```

### 3.1 아이템간의 유사도를 계산  
영화간의 피어슨 유사도를 계산

```python
movie_similarity = np.corrcoef(movie_user_matrix)
```

```python
movie_similarity = pd.DataFrame(
    data=movie_similarity,
    index=movie_user_matrix.index,
    columns=movie_user_matrix.index
)
```

```python
movie_similarity.shape
```

## 3.2 아이템 i와 비슷한 아이템을 k개 찾음  
영화 "648"과 유사한 다른 영화 k개를 찾음  
우선 movie_i와 다른 영화 간의 유사도 추출

```python
movie_i_similarity = movie_similarity.loc[movie_i]
```

다른 영화와의 유사도를 내림차순 정렬

```python
movie_i_similarity = movie_i_similarity.sort_values(ascending=False)
```

유사도 상위 k개의 유사도와 id를 추출  
이 때 가장 유사도가 높은 idsms movie_i로 제외

```python
top_k_similarity = movie_i_similarity[1: k+1]
top_k_similar_movie_ids = top_k_similarity.index
```

```python
top_k_similarity
```

### 3.3 아이템 i의 평균 선호도를 계산  
영화별로 특징이 된든 평균 선호도를 계산  
평점이 0인 경우 평가하지 않음을 반영하기 위해 결측값으로 대체

```python
movie_user_matrix = movie_user_matrix.replace(0, np.NaN)
```

```python
movie_bias = movie_user_matrix.mean(1)
```

```python
movie_bias
```

### 3.4 유저가 평가한 K개의 아이템의 선호도의 편차를 유사도 기준으로 가중 평균

#### 3.4.1 유저별 영화 평가 점수 편차 계산

```python
movie_user_matrix_wo_bias = movie_user_matrix.sub(movie_bias, axis=0)
```

```python
movie_user_matrix_wo_bias
```

#### 3.4.2 상위 k개의 선호도 추출

```python
top_k_similar_ratings = movie_user_matrix_wo_bias.loc[top_k_similar_movie_ids, user_id]
top_k_weighted_ratings = top_k_similar_ratings * top_k_similarity
```

```python
top_k_similar_ratings
```

```python
top_k_weighted_ratings
```

추출된 영화 중 평가 점수가 있는 영화에 대한 가중치만 남김

```python
top_k_weight = (pd.notna(top_k_similar_ratings)) * top_k_similarity
top_k_weight
```

#### 3.4.3 가중 평균
유사도가 곱해진 평가 점수의 편차 합을 유사도 합으로 나눔

```python
weighted_rating = top_k_weighted_ratings.sum()
weight = top_k_weight.sum()
```

```python
weight
```

영화 평균 평점 추출

```python
bias = movie_bias.loc[movie_i]
```

```python
bias
```

```python
if weight != 0:
  # 평균 평점에 가중 편차 합
  prediction_rating = bias + weighted_rating / weight

# weight가 0인 경우 유사 영화 모두 평가하지 않은 경우
else:
  prediction_rating = 0
```

```python
prediction_rating
```

### 3.5 예측 선호도가 높은 아이템을 유저에게 추천  
모든 영화에 대해서 점수를 예측하고 예측 평가 점수가 높은 영화를 유저에게 추천

```python
prediction_dict = {}

# 모든 영화 아이디에 대해 평점 예측
for movie_id in movie_user_matrix.index:
  # 이미 유저가 평가한 경우 제외
  if movie_user_matrix.loc[movie_i, user_id] > 0:
    continue

  top_k_similar_ratings = movie_user_matrix_wo_bias.loc[top_k_similar_movie_ids, user_id]

  top_k_weighted_ratings = top_k_similar_ratings * top_k_similarity
  top_k_weight = (top_k_similar_ratings != 0) * top_k_similarity

  weighted_rating = top_k_weighted_ratings.sum()
  weight = top_k_weight.sum()

  bias = movie_bias.loc[movie_i]

  if weight > 0:
    prediction_rating = bias + weighted_rating / weight
  else:
    prediction_rating = 0

  # 영화 아이디 별로 예측 평가 점수 저장
  prediction_dict[movie_id] = prediction_rating
```

영화 아이디별 예측 평가 점수를 내림차순으로 정렬

```python
prediction = pd.Series(prediction_dict).sort_values(ascending=False)
```

예측 평가 점수 상위 10개의 영화 아이디 추출

```python
recommend = prediction[:10].index
```

```python
movies.loc[recommend]
```
