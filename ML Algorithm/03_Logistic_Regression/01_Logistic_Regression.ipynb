# Logistic Regression

## Logistic Regression 정의

### 데이터의 형태
1. 연속형 데이터
  - 키, 몸무게 등
  - Liner Regression 사용 가능
2. 비연속형 데이터(범주형)
  - 성별: 남/여 등 (0, 1로 분류)
  - Linear Regression 사용 어려움
    - 비연속형 데이터의 경우 정답의 범위가 0 or 1임
    - 단순 선형 회귀의 예측값의 범위는 $-\infty$ ~ $\infty$  
    - 0과 1 사이를 벗어나는 예측은 예측의 정확도를 낮추게 되므로 예측의 결과가 0과 1 사이에 있어야 함  


-> 비연속형 데이터를 예측하고 싶은 경우 Logistic Regression을 사용할 수 있으며 이 때 Logistic Function을 사용해 예측 범위를 0과 1 사이로 제한할 수 있음  
> Logistic Function 수식
> $y = {1 \over 1 + e^{-x}}$

### Logistic Regression 이란
- Logistic Regression = Linear Regression + Logistic Function   
- 정답이 범주형일 때 사용하는 Regression 모델

**Logistic Regression 수식**
- Liner Regression
  - $y = \beta_{0} + \beta_{1}x$
- Logistic Function
  - $y = {1 \over 1 + e^{-x}}$
- Logistic Regression
  - $P(y=1) = {1 \over 1 + e^{-(\beta_{0} + \beta_{1}x)}}$

## Area Under ROC (AUROC)  
정확도는 Threshold에 따라 변하기 떄문에 평가 지표로서 부족한 경우가 있음  
이를 보완하기 위한 Threshold에 의해 값이 변하지 않는 지표가 AUROC

### ROC Curve
Roc Curve는 True Positive Ration 대 False Positive Ratio의 그래프임

#### Confusion Matrix
Confusion Matrix는 실제 값과 예측 값을 Matrix 형태로 표현한 것  
image.png

**True Positive Ratio(TPR)**  
= $True Positive \over True Positive + False Negative$

**False Positive Ratio(FPR)**  
= $False Positive \over False Positive + True Negative$

<br/> 

**Confusion Matrix 예시** 

|예측값|정답|분류|
|---|---|---|
|1|1|True Positive|
|1|1|True Positive|
|1|0|False Positive|
|0|1|False Negative|
|0|0|True Negative|
|0|0|True Negative|

위와 같은 상황에서 Confusion Matrix 값  

||Positive|Negative|
|---|---|---|
|True|2|1|
|False|1|2|


image.png

사용할 수 있는 best threshold를 구하고 싶을 때는 Youdens' Index라는 값을 사용  
$J = {True Positive \over True Positive + False Negative} + {True Negative \over True Negative + False Positive} - 1$

J값을 가장 크게해줄 수 있는 값이 best threshold

## Multiclass Logistic Regression

범주가 3개 이상인 경우  
- e.g) 혈액형 A, B, O, AB -> 0, 1, 2, 3  

### One or Nothing  
하나의 범주에 대해 해당 범주에 해당되는 지 여부를 0 혹은 1로 나타내는 방법  
- e.g. A, B, O, AB
  - A or Not
  - B or Not
  - O or Not
  - AB or Not

이 방법을 이용하면 모든 범주에 대해 해당 범주에 포함되는 것과 그 외의 것으로 나누어 아래와 같이 구분 그래프를 만들 수 있게 됨  
image.png

위 그래프를 확률로 표현하면 아래와 같음  
$P(y=0) = {1 \over 1 + e^{-(\beta_{00}+\beta_{01}x)}}$  
$P(y=0) = {1 \over 1 + e^{-(\beta_{10}+\beta_{11}x)}}$  
$P(y=0) = {1 \over 1 + e^{-(\beta_{20}+\beta_{21}x)}}$  

<br/> 

예를 들어 각 확률이 아래와 같음  
$P(y=0) = 0.3, P(y\neq0) = 0.7$  
$P(y=1) = 0.5, P(y\neq0) = 0.5$  
$P(y=2) = 0.2, P(y\neq0) = 0.8$  

확률이 제일 높은 class가 예측값
