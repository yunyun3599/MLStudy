# Non-Hierarchical Clustering

## 1. 비계층적 군집화 정의
- 전체 데이터를 확인하고 특정한 기준으로 데이터를 동시에 구분함  
- 각 데이터들은 사전에 정의된 개수의 군집 중 하나에 속하게 됨  

### 비계층적 군집화 종류
1. K-Means
2. DBSCAN

## 2. K-Means
### K-Mean 정의
- 주어진 데이터를 K개의 군집으로 묶는 방법

### K-Mean의 특징
- 각 군집은 하나의 중심을 가짐
- 각 데이터는 가장 가까운 중심에 할당
- 같은 중심에 할당된 개체들이 모여 하나의 군집을 형성
- 사전에 군집의 수, K가 정해져 있어야 함  

### K-Means 학습 방법
1. 데이터 중 임의로 K개의 중심점(Centroid)을 설정  
2. 모든 데이터에서 설정된 각 군집의 중심점까지의 거리를 계산
3. 모든 데이터를 가장 가까운 중심점이 속한 군집으로 할당
4. 각 군집의 중심점을 재설정
5. 2 ~ 5를 군집의 중심점이 변경되지 않을 때까지 반복

### K-Means의 성능에 영향을 미치는 요인
1. 초기 중심점 설정
  - K-Means의 성능은 초기 중심점의 위치에 크게 좌우됨
  - 무작위로 초기 중심이 설정되는 위험을 방지하기 위한 K-means 설정 방법  
     1. 반복적으로 수행하여 가장 많이 나타나는 군집을 사용하는 방법
     2. 전체 데이터 중 일부만 추출하여 계층적 군집화를 수행한 뒤 초기의 군집 중심을 설정하는 방법  
     3. 데이터 분포의 정보를 사용하여 초기 중심을 설정하는 방법
2. K의 개수
  - 군집의 개수 K가 어떻게 성능에 영향을 미치는 지 확인  
    image.png
  - K가 작을 때보다 클 때 과적합될 확률이 더 높음
  - K의 개수를 정하기 위한 수식 = **SSE (Sum of Squared Error)**
    - 관측치와 중심들 사이의 거리
    - $SSE = \sum_{i=1}^{K}\sum_{x \in C_i} dist(x, c_i)^2$

## 3. DBSCAN
DBSCAN = Density-Based Spatial Clustering of Applications with Noise  
> 점 P에서부터 거리가 e(epsilon) 내에 m(minPts)개 이상 있으면 하나의 군집으로 인식하는 방법  

### DBSCAN 학습 방법
1. Core Points
  - 거리 eps 이내에 데이터가 (자신 포함) minPts개 이상 있는 포인트
  - 예시) minPts = 4일 때  
    image.png  
      -> 3번은 CorePoint가 될 수 있음  
      이처럼 ccore points들을 다 계산하면 3, 4, 5, 6, 7, 8이 Core Point가 될 수 있음

2. Border Points
  - Core Points를 이웃으로 갖고 있지만 eps 이내에 데이터가 minPts개 보다 적은 포인트
  - 예시 minPts = 4  
    image.png  
       2번, 9번이 Border Point
3. Noise Points
   - Core Points를 이웃으로 갖고 있지 않고 eps 이내에 데이터가 minPts개 보다 적은 포인트
   - 예시 minPts = 4
    image.png   
      1번이 Noise Point

위의 예시에서는 1번을 제외한 나머지 포인트들을 군집으로 얻게 됨  
image.png

### DBSCAN 모수  
1. minPts
  - 최적의 minPts를 구하는 방법에 대해서는 알려진 방법이 없음  
  - Cross Validation 등의 방법으로 구해야 함
2. eps
  - 주어진 minPts에 대한 최적의 거리를 구하는 방법에 대해서는 K-dist Graph를 사용  

### K-Dist Graph
- minPts번째 인접한 이웃 데이터 포인트까지의 거리
- 급격히 증가하기 직전의 지점을 eps로 설정

K-Dist 그래프를 구한 예시 그림

### DBSCAN의 장점  
- 지역적 패턴이 있는 데이터의 군집도 찾을 수 있음  
- 노이즈 데이터를 따로 분류하여 노이즈 데이터들이 군집에 영향을 주지 않음

### DBSCAN의 단점  
- 밀도가 구역에 따라 바뀔 경우를 파악하지 못함
