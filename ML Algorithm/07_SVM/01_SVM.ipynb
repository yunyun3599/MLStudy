# SVM
## SVM의 정의
SVM = Support Vector Machine  
결정 경계(Decision Boundary)를 정의하는 모델

SVM의 목표  
- Margin을 최대화하는 Decision Boundary(결정 경계) 찾기
- Support Vector가 Decision Boundary를 만드는데 영향을 주기 때문에 Support Vector Machine이라고 불림

### Support Vector Machine의 구성요소
- Support Vector
  - 두 클래스 사이의 경계에 위치한 데이터 포인트들
- Margin
  - Decision Boundary와 서포트 벡터 사이의 거리
- Decision Boundary
  - 데이터를 나누는 기준이 되는 경계

## Margin의 종류
아래와 같이 다른 종류의 데이터가 다른 분류의 데이터 군집과 가까이 있어 SVM을 통해 분류할 때 해당 데이터의 오분류를 허락할지 말지에 따라 Hard Margin과 Soft Margin으로 나눌 수 있음  
image.png


### Hard Margin
어떠한 오분류도 허용하지 않음

### Soft Margin
어느 정도의 오분류를 허용하나 오분류시 Penalty를 부여
- Penalty의 종류
  1. 0-1 Loss
    - Error가 발생한 개수만큼 패널티 계산
    - Support Vector Machine 너머에 있는 데이터 개수에 대해 Penalty 부여
      image.png
  2. Hinge Loss
    - 오분류 정도에 따라 패널티 계산 (ERROR의 크기 반영)
      image.png

직선만으로 Hard Margin을 만족시키며 데이터의 분류를 할 수 없는 경우를 위해 **Non-Linear SVM**이 등장  

### Non Linear SVM
직선으로 데이터를 나눌 수 없을 때 사용하는 방식

## Kernel
### Kernel Trick
저차원 데이터를 고차원 데이터로 맵핑(Mapping)하는 작업

### 커널의 종류
- 선형
- 다항식
- 가우시안 RBF

#### 가우시안 RBF
다음 그림처럼 가운데에 데이터가 몰려있을 때 데이터를 직선으로 나눌 수 없으므로 데이터를 위로 볼록하게 하여 맵핑시킨 후 면을 통해 분류 가능

## SVM의 장단점
장점
- 비선형 분리 데이터도 커널을 사용하여 분류할 수 있음  
- 고차원 데이터에도 사용할 수 있음  

단점
- 데이터가 너무 많으면 속도가 느림
- 확률 추정치를 제공하지 않음
