# 머신러닝이란

## 머신러닝의 정의

기계가 일일이 코드로 명시하지 않은 동작을 데이터로부터 학습하여 실행할 수 있도록 하는 알고리즘을 개발하는 연구 분야  

-> 풀고자하는 문제의 정답과 데이터를 주고 기계를 학습시켜서 정답을 맞추게 하는 것

## 머신러닝의 3요소

1. 데이터
  - 객관적인 기준을 위한 정답과 요인들
  - 주관적인 기준이 아닌 수치에 근거한 객관적인 기준
2. 학습
  - 데이터로부터 최적의 값을 찾는 과정
  - 수치에 들어갈 값을 컴퓨터가 스스로 찾음
3. 알고리즘
  - 문제를 풀기 위한 의사 결정 과정

## 머신러닝의 종류

### 머신러닝 알고리즘의 종류
1. 지도학습
  - supervised learning
  - 정답이 있는 문제를 풀 경우
2. 비지도학습
  - unsupervised learning
  - 정답이 없는 문제를 풀 경우

### 정답의 종류에 따른 지도 학습
1. 회귀
  - Regression
  - 정답이 연속형 변수일 때
2. 분류
  - Classification
  - 정답이 비연속형(범주형) 변수일 때

#### 회귀  
- 정답이 연속형 변수일 때  
- 연속형: 값이 정수처럼 명확하지 않음 (e.g. 키, 몸무게 등)
- 회귀 분석 문제 예시: 키를 이용해 몸무게 예측

회귀 분석의 경우 데이터를 가지고 점들을 예측하는 하나의 선을 만들고, 그 선을 통해 예측값을 도출해냄

#### 분류  
- 정답이 비연속형(범주형) 변수일 때
- 범주형 변수: 값이 정수처럼 명확한 경우 (e.g. 성별, 도시 등)
- 분류 문제 예시: 키와 몸무게를 이용해 성별 예측

### 비지도 학습의 종류
1. 군집 분석
  - Clustering
  - 주어진 데이터가 어떻게 구성되었는지 알아내려는 분석 방법

2. 강화학습
  - Reinforcement Learning
  - 행동에 따른 보상을 최대화시키는 학습 방법
