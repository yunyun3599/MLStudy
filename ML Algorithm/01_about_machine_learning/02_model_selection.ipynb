# Model Selection

## 모델

### 모델의 정의
- 어떤 X가 주어졌을 때 f 라는 함수를 통해 y라는 값을 도출하는 과정
- 이 때의 함수 f를 모델 또는 알고리즘이라고 부름

> y = f(X)  
> X = 데이터  
> y = 예측값  
> f = 모델  

### 모델의 목적
- 데이터를 이용해 값을 예측

### 모델의 평가
- 모델이 값을 잘 예측하는지 평가

## 데이터의 종류
1. Train Data
  - 학습에 사용되는 데이터
2. Test Data
  - 학습에 사용되지 않는 데이터
  - 모델이 실제로 잘 예측하는 지 알기 위해서 학습에 사용되지 않은 데이터를 사용해 평가를 진행

## 과대적합, 과소적합
모델의 평가와 데이터의 관계에 대한 부분

1. Overfitting
  - 과대 적합
  - 학습 데이터(Train data)는 잘 맞추지만 학습 데이터 외에는 잘 맞추지 못하는 현상

2. Underfitting
  - 과소 적합
  - 학습 데이터(Train data)를 잘 맞추지 못하는 현상


image.png

#### Underfitting을 확인하는 방법
- Train data로 학습된 모델을 Train data로 평가 
- Train data를 잘 맞추지 못한다면 Underfitting 상태

#### Overfitting을 확인하는 방법
- Train data를 잘 학습한 모델을 Test data로 평가
- Train data는 잘 맞추지만 Test data를 잘 맞추지 못한다면 Overfitting 상태

<br/> 
#### Data Split
주어진 데이터 셋을 Train Data와 Test Data로 나눔  
- 주로 70%, 30% 수준으로 나눔
- 종종 Validation Data도 필요한 경우가 있음

<br/>  
#### 데이터의 용도
1. Train Data
  - 학습에 사용되는 데이터
2. Valid Data
  - 학습이 완료된 모델을 검증하기 위한 데이터
  - 학습에 사용되지는 않지만 관여하는 데이터
3. Test Data
  - 최종 모델의 성능을 검증하기 위한 데이터
  - 학습에 사용되지 않으며 관여하지도 않는 데이터

**Valid data도 학습에 관여하기 때문에 Valid data에 overfitting되는 경우도 발생 가능**  
-> 이 경우를 방지하기 위해 **Cross Validation**을 정의

## Cross Validation
**교차 검증**  
Valid 데이터를 고정하지 않고 계속해서 변경함으로써 Overfitting 되는 것을 막기 위한 방법

### Cross Validation 종류
1. LOOCV (Leave One Out Cross Validation)
  - 하나의 데이터를 제외하고 모델을 학습한 후 평가
  - image.png
  - 데이터를 하나씩 빼고 학습 후 해당 데이터를 가지고 검증하는 과정을 모든 데이터에 대해 수행
  - 데이터 개수만큼 모델을 학습해야함
  - 데이터가 많을 경우 시간이 오래 걸림
2. K-Fold
  - 데이터를 K개로 분할한 후 한 개의 분할 데이터를 제외한 후 학습에 사용
  - 제외된 데이터를 학습이 완료된 후 평가에 사용
  - 3-Fold 예시
    - image.png

#### Cross validaion 평가
1. Cross Validation을 이용하면 방법에 따라 K개의 평가 지표가 생성
2. 생성된 평가 지표의 평균을 이용해 모델의 성능을 평가
3. 전체 Train 데이터를 이용해 모델 학습


가장 좋은 성능을 보이는 모델로 전체 Train 데이터를 이용해 모델 학습
