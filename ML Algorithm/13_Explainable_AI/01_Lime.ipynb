# Explainable AI

## 설명 가능한 인공지능의 정의
어떤 알고리즘이 의사결정을 내리는 과정을 설명하기 위한 방법

<br/>

### 설명 가능한 인공지능(Explainable AI)의 필요성  

||설명 가능성|정확도|
|:--:|:--:|:--:|
|복잡한 모델|X|O|
|간단한 모델|O|X|

위와 같은 상황을 해결하기 위해서 아래 두가지 접근법을 시도해볼 수 있음  
1. 간단한 모델의 성능을 높이는 방법
2. 복잡한 모델을 설명하는 방법  

그러나 간단한 모델의 성능을 높이는 데는 한계가 있기 때문에 최근에는 복잡한 모델을 설명하는 접근 방식을 많이 쓰고 있음  

### 설명이 가능한 머신러닝 알고리즘 종류  
1. Linear Regression
  - 예시: $집값 = 0.3 \times 평수 - 0.2 \times 역과의 거리$
    - 평수가 1 증가하면 집값은 0.3 증가  
    - 역과의 거리가 1 증가하면 집값은 0.2 감소
    - 30평이고 역과의 거리가 10인 집값은 7
2. Decision Tree
  - 예시  
    - image.png
    - 집이 30평 이상이고 역과의 거리가 10 이상이면 집값은 6

### 설명이 어려운 머신러닝 알고리즘 종류  
1. Random Forest
  - Random Forest의 경우 여러 개의 나무로 이루어져있으나, 그 중 특정 하나의 나무가 전체를 대변하지 않음  
  - 따라서 어떤 변수가 중요한 지는 알지만 이 변수가 의사결정에 어떻게 영향을 끼치는 지는 알 수 없음

## Lime
- Local Interpretable Model-agnostic Explanations
- 모델이 단일 Sample을 왜 그 예측을 했는지 해석하는 모델  

image.png


### Local
- Lime은 모델 전체에 대한 설명보다는 모델이 단일 Sample을 왜 그렇게 예측했는지 해석
- 데이터 공간의 전체에 대한 모델의 설명은 어려움  
- 하지만 설명하려는 데이터 근처에서는 의미 있는 모형으로 설명 가능  

### Model-agnositc
- Lime은 모델에 의존적이지 않음  
- 어떤 모델에도 적용 가능
