# Anomaly Detection

## 이상 탐지(Anomaly Detection)의 정의
- 주어진 데이터의 정상 여부를 판별하는 문제
- 신용 카드 사기, 대출 사기 탐지 등이 대표적인 예


 ### 이상 탐지의 특징
 - 이상 탐지 문제의 가장 큰 특징은 정상 데이터와 이상 데이터의 비율의 불균형

## 데이터 불균형
### 데이터 불균형(Class Imbalance)의 정의  
- 지독 학습에서 모델을 학습하기 위해서는 정답이 있는 데이터가 필요  
- 예를 들어 Binary Classification 문제에서 한 쪽의 데이터가 매우 적은 상황  
  - 데이터가 많은 그룹 = Majority Group
  - 데이터가 적은 그룹 = Minority Group

### 데이터 불균형의 문제  
- 모델이 모든 예측 값을 데이터가 많은 쪽(Majority Group)으로 예측하는 상황이 발생  
- Decision Tree의 경우 모든 데이터를 Majority Group으로 예측하면 더 많은 정보 이득을 얻을 수 있게 됨  

예시
  - Class 1: Majority Group 데이터 1000개
  - Class 2: Minority Group 데이터 1개  
  - 위의 경우 예측값을 항상 Class 1로 분류하면 정확도는 ${1000 \over 1001} = 0.99$ 가 되어 높은 정확도를 갖게 됨  


### 데이터 불균형의 해결 방법
1. 샘플링을 통한 해결 방법
  - Under Sampling
  - Over Sampling
2. 모델을 통한 해결 방법

## Under Sampling
### Under Sampling의 정의
- Majority Group의 데이터를 덜 뽑아서 데이터의 비율을 맞추는 방법  

### Under Sampling의 방법
1. Random Under Sampling
  - Majority Group의 데이터를 무작위로 삭제하는 방법  
  - 장점: 빠르고 쉬움
  - 단점: Majority Group의 정보를 잃을 수 있음
2. Near Miss Under Sampling
  - Minority Group 근처에 있는 Majority Group의 데이터를 선택하는 방법  
  - Near Miss Under Sampling 방법
    1. Near-Miss ver.1
      - 가장 가까운 3개의 Minority Group 데이터의 평균 거리
    2. Near-Miss ver.2
      - 가장 먼 3개의 Minority Group의 데이터의 평균 거리
    3. Near-Miss ver.3
      - 가장 가까운 Minority Group의 데이터와의 거리


## Over Sampling
- Minority Group의 데이터를 더 뽑아서 데이터의 비율을 맞추는 방법

### Over Sampling의 방법  
1. Simple Over Sampling
  - Minority Group에서 데이터를 더 많이 추출하는 방법  
  - image.png
  - 장점
    - 빠르게 적용할 수 있음
  - 단점
    - 데이터의 새로운 정보를 제공하지 않음  
    - 과대 적합 될 가능성이 큼


2. SMOTE (synthetic minority oversampling technique)
  - 합성 데이터를 생성해서 데이터의 비율을 맞추는 방법
  - SMOTE 방법
    1. Minority Group에서 임의의 데이터를 선택  
    2. KNN으로 선택된 데이터와 Minority Group 중 가까운 이웃을 찾음
    3. 선택된 데이터와 가까운 이웃 사이의 거리를 측정함  
    4. 0~1 사이의 임의의 수를 곱하여 데이터를 생성함  
    5. 데이터의 비율이 같아질 때까지 1~4를 반복

## 평가 방법
### Anomaly Detection과 Accuracy
- 정답을 Majority Group으로만 예측할 경우 Accuracy는 상승
- Minority Group을 잘 예측하는지 알 수 없다는 단점이 있음  

### Accuracy 계산 방법  
$Accuracy = {True Positive + True Negative \over All Data}$    
image.png

### F1-Score
- 정확도는 잘못 분류된 데이터를 반영하지 않는 문제가 있음  
- F1 Score는 Precision과 Recall의 조화 평균으로 계산됨  
- 잘못 분류된 데이터를 반영한 평가지표

#### Precision 계산 방법  
$Precision = {True Positive \over True Positive + False Positive}$   
image.png

#### Recall 계산 방법  
$Recall = {True Positive \over True Positive + False Negative}$  
image.png

#### F1-Score 계산 방법  
$F1 = 2* {(Precision * Recall) \over Precision + Recall}$

### 아래 경우에 각 지표 계산해보기  
image.png

**Precision**  
$Precision = {10 \over 10 + 10} = 0.5$  

**Recall**  
$Recall = {10 \over 10 + 10} = 0.5$  

**F1-Score**  
$F1 = 2 * {(0.5 * 0.5) \over 0.5 + 0.5} = 0.5$  

**Accuracy**  
$Accuracy = {(10 + 1000) \over 10+10+10+1000} = 0.98$

<br/>

**Accuracy vs F1-Score**
- Accuracy = 0.98
- F1 = 0.5
- 높은 수치인 Accuracy와 다르게 F1은 잘못 예측된 데이터를 반영 가능
