# Out-of-Distribution

## Out-of-Distribution 정의
- 모델은 정상 범위의 데어트 (In-Distribution)만을 학습
- 정상 범위에서 먼 데이터(Out-of-Distribution)을 이상으로 판단하는 이상 탐지 모델


### Out-of-Distribution 종류
1. Isolation
2. OCSVM
3. PCA

## OCSVM
- One-Class SVM
- 데이터를 Mapping한 뒤 정상 데이터를 원점에서 멀어지게 하는 방법  
- Feature Space에서 원점에서 가까운 데이터는 비정상 데이터
- Feature Space에서 원점에서 먼 데이터는 정상 데이터
