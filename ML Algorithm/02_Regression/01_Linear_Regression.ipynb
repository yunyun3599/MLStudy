# Linear Regression

## 선형 회귀 정의

예측값을 직선으로 표현하는 모델  
데이터를 잘 예측하는 선을 구하는 것  

### 선형 회귀 수식
$\widehat{y} = B_{0} + B_{1}x$

### 데이터  
$\widehat{y}$: 예측값 (prediction)  
$x$: 입력값 (Data)

### 계수
$B_{0}$: 편향  
  - 그래프에서 회귀선의 위치
  - x가 0일때의 예측값

$B_{1}$: 계수 (Coefficient)
  - 회귀선의 기울기
  - x가 1 증가할 때 예측값이 증가하는 정도


## 선형 회귀 해석

## 선형 회귀 평가
더 잘 예측하는 직선 = 실제값과 예측값의 차이가 작은 직선

```
X = (1, 2, 3, 4)  # 입력 데이터
Y = (2, 1, 4, 3)  # 실제 정답
```

**직선 1**  
$y = 1 + {1 \over 2}x \to \hat{Y} = (1.5, 2, 2.5, 3)$
> Y - $\hat{Y}$ = (0.5, -1, 1.5, 0)

**직선 2**  
$y = 1 + x \to \hat{Y} = (1, 2, 3, 4)$
> Y - $\hat{Y}$ = (1, -2, 1, -1)

<br/>. 
Y - $\hat{Y}$ 값들을 단순히 더하면  
결과가 (0, 0, 0, 0)으로 모두 정답을 맞춘 경우와  
결과가 (1, -1, 1, -1)으로 정답을 맞추지 못한 경우의 평가 지표 값이 동일하게 나오는 문제가 발생  



### Mean Squared Error
위와 같은 문제를 막기 위해 차이의 크기를 반영하기 위해 제곱 후 평균을 구하는 방식

Y - $\hat{Y}$ = (0, 0, 0, 0) -> ${1 \over n} \Sigma(Y-\hat{Y})^2 = 0$   
Y - $\hat{Y}$ = (1, -1, 1, -1) -> ${1 \over n} \Sigma(Y-\hat{Y})^2 = 1$
