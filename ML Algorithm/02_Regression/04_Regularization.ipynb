## Linear Regression & Regularization
선형 회귀의 목표 = 실제 값을 잘 예측하는 것 (Loss 최소화)
> $argmin(y - βX)^2$

Regularization을 더해 불필요한 beta는 학습하지 않도록 하는 것  
> $argmin(y-βX)^2 + λΣ||β||_{i}$  

$||β||_{i}$: 계수들의 Norm   
<br/>  

**$||β||_{i}$ 계산 방법**  
L1 Norm: $||β||_{1} = \sum_{i=1}^N |β_{i}|$  
L1 Norm: $||β||_{2} = \sum_{i=1}^N (β_{i})^2$

## 선형회귀 + 정규화의 종류
1. LASSO (least absolute shrinkage and selection operator)
   - L1 norm: $argmin(y-βX)^2 + λΣ||β||_{1}$
2. Ridge
  - L2 norm: $argmin(y-βX)^2 + λΣ||β||_{1}$
