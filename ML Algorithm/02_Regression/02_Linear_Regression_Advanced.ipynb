# Linear Regression 심화

## Multivariate Regression
다변량 회귀
- 두 개 이상의 변수로 만들어진 회귀식

e.g. 키  = b0 + b1\*몸무게 + b2\*나이

<br/>  

- 변수가 2개인 경우
  - $y = \beta_{0} + \beta_{1}x_{1} + \beta_{2}x{2}$
- 변수가 여러개인 경우
  - 행렬로 표시
- 변수가 3개인 경우
  - $y = \beta_{0} + \beta_{1}x_{1} + \beta_{2}x{2} + \beta_{3}x{3}$
  - $y = \begin{bmatrix} y1 \\ y2 \\ y3 \\ y4\end{bmatrix} \beta = \begin{bmatrix} \beta_{0} \\ \beta_{1} \\ \beta_{2} \\ \beta_{3}\end{bmatrix} X=\begin{bmatrix} 1 && x_{11} && x{12} \\ 1 && x_{21} && x_{22} \\ 1 && x_{31} && x_{32} \\ 1 && x_{41} && x_{42} \end{bmatrix}$ => $y = \beta X$
    - X 행렬의 1은 편향 $\beta_{0}$를 살리기 위해 있음

$y = 1 + {1 \over 2}x_{1} + 2x_{2}$ 
- $x_{1}$이 1 증가할 때 예측값은 0.5 증가
- $x_{2}$이 1 증가할 때 예측값은 2 증가
- $x_{2}$가 $x_{1}$보다 더 큰 영향을 미침

### 다항식 회귀 vs 다변량 회귀

1. 다항식 회귀
  - 새로운 변수가 추가되는 것이 아니라 기존 변수의 n제곱 형태가 추가됨
  - $y = \beta_{0} + \beta_{1}x + \beta_{2}x^2$
2. 다변량 회귀
  - $y = \beta_{0} + \beta_{1}x + \beta_{2}x_{2}$

### 비션형 표현 방식
1. 비선형 회귀식
  - $y = \beta_{0} + \beta_{1}x + \beta_{2}x^2$
2. 완전제곱식으로 표현
  - $y = a(x-b)^2 + c$

polynomial regression은 x제곱, 3제곱, 4제곱등도 표현 가능

## 회귀 계수를 계산하는 방법

## 통계적 방법
$y = \beta_{0} + \beta_{0}x_{1} + \beta_{0}x_{2}$  
$\boldsymbol{y} = \boldsymbol{\beta X}$  

최소제곱법으로 구한 $\boldsymbol{\beta}$ Equation
=> $\boldsymbol{\beta} = (X^TX)^{-1}X^Ty$  
=> msa가 최소가 되는 $\boldsymbol{\beta}$값을 구할 수 있게 됨  

### 통계적 방법 예시
식: $y = \beta_{0} + \beta_{1}x$  
X = (1, 2, 3, 4)  
Y = (2, 1, 4, 3)

따라서  
$\beta = \begin{bmatrix} \beta_{0} \\ \beta_{1} \end{bmatrix} $
$X = \begin{bmatrix} 1 && 1 \\ 1 && 2 \\ 1 && 3 \\ 1 && 4 \end{bmatrix}$
$y = \begin{bmatrix} 2 \\ 1 \\ 4 \\ 3 \end{bmatrix}$  
이므로  
$X^T = \begin{bmatrix} 1 && 1 && 1 && 1 \\ 1 && 2 && 3 && 4 \end{bmatrix}$
$X^TX = \begin{bmatrix} 4 && 10 \\ 10 && 30 \end{bmatrix}$
$(X^TX)^{-1} = \begin{bmatrix} 1.5 && -0.5 \\ -0.5 && 0.2 \end{bmatrix}$

$\beta$를 구하면  
$\boldsymbol{\beta} = (X^TX)^{-1}X^Ty$  
$\boldsymbol{\beta} = \begin{bmatrix} 1.5 && -0.5 \\ -0.5 && 0.2 \end{bmatrix}\begin{bmatrix} 1 && 1 && 1 && 1 \\ 1 && 2 && 3 && 4 \end{bmatrix}\begin{bmatrix} 2 \\ 1 \\ 4 \\ 3 \end{bmatrix}$  
$\boldsymbol{\beta} = \begin{bmatrix} 1 && 0.5 && 0 && -0.5 \\ -0.3 && -0.1 && 0.1 && 0.3 \end{bmatrix} \begin{bmatrix}2 \\ 1 \\ 4 \\ 3 \end{bmatrix}$  
$\boldsymbol{\beta} = \begin{bmatrix}1 \\ 0.6 \end{bmatrix}$  
<br/>  
따라서  
**=> $y = 1 + 0.6x$**  

좋은 선형 회귀식을 찾는 방법  
= 여러 값을 넣어본 뒤 Loss(MSE)가 제일 작은 $\beta$를 찾기  

효율적으로 여러 값을 넣는 방법  
= 최적화 알고리즘
1. Bisection Method
2. Gradient Descent

### Bisection Method
1. 임의의 두개의 값을 설정
2. 두 값의 y값을 비교
3. y 값이 큰 점을 두 점의 가운데 점으로 바꿈
4. 임의의 두 값의 차이가 작아질 때까지 1~3을 반복

### Gradient Descent (경사 하강법)
함수의 기울기(경사)를 구하고 경사의 절댓값이 낮은 쪽으로 계속 이동시켜 극값에 이를 때까지 반복시키는 방법  

**수식**  
$x_{i+1} = x_{i} - \gamma∇f(x_{i})$   

**방법**  
1. 임의의 값 하나를 설정
2. 임의의 값에서 기울기를 계산
3. 기울기와 Learning rate를 곱한 값을 임의의 값에서 뺌
4. 기울기가 0에 가까워질 때까지 1~3을 반복
