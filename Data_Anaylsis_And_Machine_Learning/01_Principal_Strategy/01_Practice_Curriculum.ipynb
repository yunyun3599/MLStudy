# 실습 목차  
1. 보험료 예측하기
  - 다양한 Regression 방법 이용
  - 모델 성능 종합 비교
2. 대출 상품 투자 위험도 줄이기
  - CatBoost
3. HR 데이터를 활용한 직원 이탈 요인 분석
  - SMOTE를 이용한 Over-Sampling
  - Random Forest, Gradient Boosting, GridSearchCV
  - Feature Importance
4. 고객 데이터를 활용한 마케팅 분석
  - Recursive Feature Elimination, SelectKBest를 적용한 Logistic Regression, SVM, RandomForest, XGBoost
  - PCA, K-Means Clustering을 이용한 Segmentation 분석
5. 마케팅을 위한 연관 규칙 분석
  - TransactionEncoder와 Apriori 알고리즘을 적용하여 연관 규칙 분석
6. 반도체 제조 공정 프로세스 Pass & Fail 수율 예측
  - PCA, 이상치 제거  
  - Classifier들과 Grid Search를 활용한 최적 모델 탐색  
7. 자동차 제조 테스트 공정 시간 예측  
  - XGBoost, Random Forest 기초 모델
  - 차원 축소 기법을 활용한 Feature Selection
  - Stacked Regression
8. 제조 공정 예측 유지 보수  
  - XGBoost
9. 가스 터빈 추진 플랜트 제조 공정의 부식 예측
  - PCA를 활용한 Feature Selection
  - 다양한 Regressor 모델의 Hypertuning
10. 제조 공정 내 가스 혼합물의 개별 가스 성분 분류
  - PCA, Recursive Feature Elimination, SelectKBest 적용  
  - 다양한 Classifier와 Voting Ensemble
11. COVID-19 관련 이커머스 상품 판매 데이터 분석
  - K-Means Clustering과 PCA를 적용한 군집 분속  
  - Logistic Regression과 SVM을 활용  
12. 온라인 쇼핑몰 고객 Clustering 분석
  - RFM 마케팅 분석
  - K-Means Clustering
  - DB-Scan Clustering
13. 소매 판매 데이터를 활용한 이커머스 고객 Segmentation 분석
  - NLTK 패키지를 활용한 자연어 처리
  - PCA, K-Means Clustering을 통한 군집 분석
  - 다양한 Classifier 및 Voting Classifier를 통한 분류 예측 분석  
14. 시계열 분석을 이용한 Drug store 매출 예측
  - 데이터에 대한 계절성, 추세, 자기 상관 기반 시계열 분석
  - Prophet 패키지를 활용한 매출 예측 분석
15. 이커머스 여름 의류 매출 분석
  - K-Means Clustering
  - 다양한 Classifier들, Voting Classifier, GridSearchCV
