## 분석 방법  
연관 규칙 분석 방법 사용  
1. 데이터 가공 (TransactionEncoder)
2. Apriori 알고리즘
3. Association Rules

### 연관 규칙 분석 절차  
1. 모든 거래에서 발생하는 모든 항목에 대한 빈도 테이블을 생성
2. support(지지도)가 임의의 값보다 큰 것들로 필터링
3. 중요 항목의 모든 가능한 조합을 만들기  
4. 모든 조합의 발생 횟수 계싼
5. Association Rules 적용
6. 해석
