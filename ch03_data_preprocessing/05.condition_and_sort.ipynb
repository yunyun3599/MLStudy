# 조건에 맞는 데이터 추출

데이터 추출

```python
import pandas as pd
```

```python
df = pd.DataFrame({'a': [i for i in range(1, 11)], 'b': [i for i in range(11, 21)], 'c': [i for i in range(21, 31)]})
df
```

a, c열 출력하기

```python
df[['a', 'c']]
```

a가 3 이상인 데이터를 출력하기

```python
df[df['a'] >= 3]
```

a가 3 이상인 데이터 중 a, c열만 출력

```python
df[df['a'] >= 3][['a', 'c']]
```

a가 3 이상이고, b가 16 미만인 데이터 출력

```python
df[(df['a'] >= 3) & (df['b'] < 16)]
```

```python
a = (df['a'] >= 3) & (df['b'] < 16)
a
```

```python
type(a)
```

a가 3 이하이거나 7 이상인 데이터

```python
df[(df['a'] <= 3) | (df['a'] >= 7)]
```

a가 3 이상이고 b가 16 미만이거나 c가 30인 데이터

```python
df[(df['a'] >= 3) & ((df['b'] < 16) | (df['c'] == 30))]
```

# 정렬

```python
df = pd.DataFrame({'a': [2, 3, 2, 7, 4], 'b': [2, 1, 3, 5, 3], 'c':[1, 1, 2, 3, 5]})
df
```

인덱스 기준 정렬

```python
df.sort_index()
```

내림차순일 때는 ascending=False

```python
df.sort_index(ascending=False)
```

```python
df
```

결과를 저장하고 싶을 때는 inplace=True

```python
df.sort_index(ascending=False, inplace=True)
df
```

인덱스 초기화는 reset_index

```python
df.reset_index()
```

```python
df.reset_index(drop=True)
```

```python
df.reset_index(drop=True, inplace=True)
df
```

값 기준 정렬

```python
df = pd.DataFrame({'a': [2, 3, 2, 7, 4], 'b': [2, 1, 3, 5, 3], 'c':[1, 1, 2, 3, 5]})
df
```

a열 기준으로 오름차순 정렬

```python
df.sort_values(by=['a'])
```

a열 기준으로 내림차순 정렬

```python
df.sort_values(by=['a'], ascending=False)
```

a, b열 기준으로 오름차순 정렬

```python
df.sort_values(by=['a', 'b'])
# a 기준으로 먼저 정렬된 후에 a 열의 값이 같은 경우 b열 값을 기준으로 정렬
```

a열 기준으로 오름차순 정렬 후 b열 기준으로 내림차순 정렬

```python
df.sort_values(by=['a', 'b'], ascending=[True, False], inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```
