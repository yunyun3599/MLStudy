# Series

### 데이터 생성

```python
import pandas as pd
```

```python
df = pd.DataFrame({'a': [1, 2, 3], 'b':[4, 5, 6], 'c': [7, 8, 9]})
df
```

a열 추출하기

```python
df['a']
```

```python
type(df['a'])
```

### 시리즈 생성 방법

시리즈: 한 줄짜리 데이터 (리스트와의 차이점은 인덱싱 번호가 있다는 것)

```python
series = pd.Series([1, 2, 3, 4, 5])
series
```

### 인덱스 변경하기

```python
indexed_series = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
indexed_series
```

```python
indexed_series['b']
```

### unique 값 찾기

```python
df = pd.DataFrame({'a': [1, 2, 3, 1, 2, 3], 'b':[4, 5, 6, 6, 7, 7], 'c': [6, 7, 8, 9, 9, 9]})
df
```

```python
df_a = df['a']
df
```

```python
type(df_a)
```

```python
df_a.unique()
```

```python
df_a.unique()[2]
```
