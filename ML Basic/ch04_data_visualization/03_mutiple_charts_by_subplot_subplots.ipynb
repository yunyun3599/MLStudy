```python
import matplotlib.pyplot as plt
```

data1(x: 1,2,3; y: 1,2,3) 과 data2(x: 1,2,3; y: 1,100,200)을 그래프로 출력

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

## subplot을 이용한 방법

`subplot(행의 수, 열의 수, 해당 그래프가 그려질 위치)`

```python
# 1행 2열의 모양으로 그림
plt.subplot(1, 2, 1)  # 1행 1열에 위치시킬 그래프 (첫번째 위치)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(1, 2, 2)  # 1행 2열에 위치시킬 그래프 (두번째 위치)
plt.plot(x2, y2)
plt.title('data2')
```

```python
# 2행 1열의 모양으로 그림
plt.subplot(2, 1, 1)  # (2, 1)의 첫번째 위치
plt.plot(x1, y1)
plt.title('data1')
plt.subplot(2, 1, 2)  # (2, 1)의 두번째 위치
plt.plot(x1, y1)
plt.title('data2')
```

tight_layout을 이용하면 레이아웃이 자동으로 설정됨

```python

plt.subplot(2, 1, 1)  # (2, 1)의 첫번째 위치
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)  # (2, 1)의 두번째 위치
plt.plot(x1, y1)
plt.title('data2')

plt.tight_layout()
```

data(x: 1,2,3 y:1,2,3)과 data2(x: 1,2,3 y:1,50,200)을 그래프로 출력

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 50, 200]
```

```python
# 한 가지 그림에 축을 2개 만들어서 표현하는 방법

fig, axe1 = plt.subplots()

axe2 = axe1.twinx()   # x축을 공용으로 쓰는 다른 축 생성
axe1.plot(x1, y1, color='red', label='data1') # 축 1
axe2.plot(x2, y2, color='blue', label='data2')  # 축 2

axe1.set_xlabel('x', fontsize=15) # x축 라벨 생성
axe1.set_ylabel('y1', fontsize=15)  # 좌측 y축 라벨 생성
axe2.set_ylabel('y2', fontsize=15)  # 우측 y축 라벨 생성
```

y축을 두 개 가진 차트에서 범례 표시하기

```python
fig, axe1 = plt.subplots()

axe2 = axe1.twinx()
chart1 = axe1.plot(x1, y1, color='red', label='data1')
chart2 = axe2.plot(x2, y2, color='blue', label='data2')

axe1.set_xlabel('x', fontsize=15)
axe1.set_ylabel('y1', fontsize=15)
axe2.set_ylabel('y2', fontsize=15)

chart = chart1 + chart2
axe1.legend(chart, ['data1', 'data2'])
```
