# linechart를 이용한 선도표 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': (10, 5)})
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
# 결측값 확인
df.isnull().sum()
```

선도표 그리기

```python
# 연도별 승객 수
sns.lineplot(data=df, x='year', y='passengers')

# 결과는 연도별로 신뢰구간을 포함해서 나오게 됨
```

xtick에 전체 년도 표시

```python
sns.lineplot(data=df, x='year', y='passengers')
plt.xticks(df['year']);   # ;를 넣지 않으면 변환하면서 발생하는 로그들이 다 출력되므로 세미콜론을 추가함
```

xtick 기울기 추가

```python
sns.lineplot(data=df, x='year', y='passengers')
plt.xticks(df['year'], rotation=45);
```

월별로 그리기

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month')
```

색상 변경

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month', palette='Set2')
```

선 표현 방법 변경

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month', palette='Set2', style='month')   # 그래프 선의 모양이 달라짐
```
