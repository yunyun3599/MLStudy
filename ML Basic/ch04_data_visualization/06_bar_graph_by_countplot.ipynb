# countplot을 이용한 막대 그래프 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': (10, 5)})
import pandas as pd
```

titanic 데이터셋의 성별 인원 수 시각화

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.groupby(by='sex')['sex'].count()
```

countplot을 이용해 시각화

```python
sns.countplot(data=df, x='sex')
```

가로로 그리기

```python
sns.countplot(data=df, y='sex')
```

titanic 데이터 셋의 성별 인원수를 객실 등급별로 시각화

```python
df.head()
```

```python
df[['sex', 'class']]
```

```python
df.groupby(by=['sex', 'class'])['sex'].count()
```

```python
sns.countplot(data=df, x='sex', hue='class')  # hue는 카테고리 대분류
```

palette를 이용한 색상 조정

```python
sns.countplot(data=df, x='sex', hue='class', palette='flare')
```
