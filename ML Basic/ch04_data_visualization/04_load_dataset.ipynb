# load_dataset을 이용한 데이터셋 불러오기

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

load_dataset 사용법: sns_load_dataset(데이터set명)  
[데이터 소스](https://github.com/mwaskom/seaborn-data)

### 예제 1: Iris 데이터

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

iris 데이터 셋의 petal_length와 petal_width를 이용하여 산점도 그리기

```python
# matplotlib

plt.scatter(df['petal_length'], df['petal_width'])
```

```python
# seaborn

sns.scatterplot(data=df, x='petal_length', y='petal_width')
```

```python
# matplotlib과 호환
sns.scatterplot(data=df, x='petal_length', y='petal_width')
plt.title('iris')
```
