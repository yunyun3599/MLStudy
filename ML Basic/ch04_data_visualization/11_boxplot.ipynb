```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': (10, 5)})
import pandas as pd
```

데이터셋 로드

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

boxplot 그리기

```python
sns.boxplot(data=df)
```

boxplot의 3개의 가로줄과 박스는 각각 0\~25%, 25\~50%, 50\~75%, 75\~100%를 의미  
가장 위 / 아래 가로줄 밖의 점들은 극단적으로 작거나 큰 이상치를 의미

세로 방향으로 그림 키우기

```python
plt.subplots(figsize=(7,8))
sns.boxplot(data=df)
```

가로방향으로 그리기

```python
sns.boxplot(data=df, orient='h')
```

swarmplot을 이용한 raw data 확인

```python
plt.subplots(figsize=(10, 10))
sns.swarmplot(data=df)
```

겹쳐 그리기

```python
plt.subplots(figsize=(10, 10))
sns.boxplot(data=df)
sns.swarmplot(data=df)
```

swarmplot 색상 조정

```python
plt.subplots(figsize=(10, 10))
sns.boxplot(data=df)
sns.swarmplot(data=df, color='black')
```

violinplot

```python
sns.violinplot(data=df) # 값의 분포대로 가로 너비가 책정되는 그래프 
```

stripplot

```python
sns.stripplot(data=df)
```
