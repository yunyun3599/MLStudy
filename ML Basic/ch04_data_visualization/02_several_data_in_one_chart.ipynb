# 여러 데이터를 한 차트에 그리기

```python
import matplotlib.pyplot as plt
```

data1(x: 1,2,3; y: 1,2,3) 과 data2(x: 1,2,3; y: 1,4,7)을 그래프로 출력

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 4, 7]
```

```python
plt.plot(x1, y1, color='blue')
plt.plot(x2, y2, color='red')
```

```python
plt.plot(x1, y1, 'b', x2, y2, 'r')
```

범례 추가 (어떤 데이터가 어떤 그래프에 해당하는 지 설명 추가)

```python
# 방법 1
plt.plot(x1, y1, color='blue', label='data1')
plt.plot(x2, y2, color='red', label='data2')
plt.legend()
```

```python
# 방법 2
plt.plot(x1, y1, color='blue')
plt.plot(x2, y2, color='red')
plt.legend(['data1', 'data2'])
```

범례 위치 / 크기 설정 ([관련문서](https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.legend.html#:~:text=Legend-,Other%20Parameters,-locstr%20or))
- 위치 설정

```python
plt.plot(x1, y1, color='blue')
plt.plot(x2, y2, color='red')
plt.legend(['data1', 'data2'], loc='upper right')
```

- 범례 폰트 사이즈 변경

```python
plt.plot(x1, y1, color='blue')
plt.plot(x2, y2, color='red')
plt.legend(['data1', 'data2'], fontsize=20)
```
