# matplotlib를 이용한 단일 차트 그리기

```python
# library 호출

import matplotlib.pyplot as plt
```

x가 1, 2, 3이고 y가 4, 5, 6인 그래프 그리기

```python
x = [1, 2, 3]
y = [4, 5, 6]
```

```python
plt.figure()
plt.plot(x, y)
```

```python
plt.plot(x, y)    # plt.figure() 생략 가능
```

lineplot은 y값만 있으면 그림을 그릴 수 있음  
이 때 x의 값은 0, 1, 2... 순서로 자동 지정됨

```python
plt.plot(y)
```

linewidth를 이용한 선 굵기 조정

```python
plt.plot(x, y, linewidth = 10)
```

color를 통한 선 색상 조정

```python
plt.plot(x, y, color='red')
```

marker를 이용한 데이터 위치 표시 [marker 관련 문서](https://matplotlib.org/stable/api/markers_api.html)

```python
plt.plot(x, y, marker = 'o')
```

marker 크기 변경

```python
plt.plot(x, y, marker='o', markersize=10)
```

선 형태 변경

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, 'o')
```

```python
plt.plot(x, y, 'ro')    # red 색상, 동그라미로 표시
```

그래프명

```python
plt.plot(x, y)
plt.title('title')
```

그래프명 폰트 크기 변경

```python
plt.plot(x, y)
plt.title('title', fontsize=20)
```

x축 명, y축 명

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

축 명 폰트 크기 변경

```python
plt.plot(x, y)
plt.xlabel('x', fontsize='20')
plt.ylabel('y', fontsize='20')
```

그리드 설정 (전체)

```python
plt.plot(x, y)
plt.grid(True)
```

그리드 설정 (x 축만)

```python
plt.plot(x, y)
plt.grid(True, axis='x')
```

그리드 설정 (y 축만)

```python
plt.plot(x, y)
plt.grid(True, axis='y')
```

x축, y축 범위 설정

```python
plt.plot(x, y)
plt.xlim([1, 2])
plt.ylim([4, 5])
```

axis를 사용한 범위 설정

```python
plt.plot(x, y)
plt.axis([1, 2, 4, 5])    # plt.axis([x의 최소값, x의 최대값, y의 최소값, y의 최대값]) 으로 설정
```

눈금 글꼴 크기 변경

```python
plt.plot(x, y)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
```

그래프에 텍스트 삽입

```python
plt.plot(x, y)
plt.text(2, 5, 'text', fontsize='20')
```

한글 폰트 사용하기

```python
plt.plot(x)
plt.title('차트명')
```
