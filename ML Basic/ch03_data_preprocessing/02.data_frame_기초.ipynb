# 컬럼명 추출 / 변경

## 데이터 프레임 생성

```python
df = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6], 'c': [7, 8, 9]})
```

```python
df
```

```python
df.columns
```

```python
df.columns[1]
```

데이터프레임의 칼럼명을 d, e, f로 변경

```python
df.columns = ['d', 'e', 'f']
```

```python
df
```

데이터 프레임의 'd' 칼럼을 '디'로, 'f'를 '에프'로 변경

```python
df.columns = ['디', 'e', '에프']
df
```

rename을 통한 칼럼명 변경

```python
df = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6], 'c': [7, 8, 9]})
df.columns = ['d', 'e', 'f']
df
```

```python
df.rename(columns = {'d': '디', 'f': '에프'})
```

```python
df
```

변경 사항 기존 변수에 저장

```python
df.rename(columns = {'d': '디', 'f': '에프'}, inplace=True) # inplace 이용
df
```

# Copy를 이용한 데이터 복사

### 데이터 프레임 생성

```python
df = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6], 'c': [7, 8, 9]})
```

```python
df.columns = ['d', 'e', 'f']
df
```

데이터 프레임 하나 더 생성

```python
df2 = df
df2
```

df의 컬럼을 다시 a, b, c로 변경

```python
df.columns = ['a', 'b', 'c']
df
```

```python
df2
```

df2의 컬럼명도 함께 변경됨 -> df, df2는 같은 객체를 참조하고 있기 때문 (얕은 복사)

해결 방법: deepcopy 이용! (깊은 복사)

```python
import copy

df2 = copy.deepcopy(df)
```

```python
df
```

```python
df2
```

```python
df.columns = ['d', 'e', 'f']
df
```

```python
df2 # df 변경사항 반영 X
```
