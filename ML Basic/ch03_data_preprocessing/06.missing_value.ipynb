# 결측값 처리

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a': [1, 1, 3, 4, 5], 'b': [2, 3, np.nan, 3, 4], 'c': [3, 4, 7, 6, 5]})
```

```python
df
```

결측 유무 확인

```python
df.isnull()
```

결측값 개수 확인

```python
df.isnull().sum()     # true는 1, false는 0으로 처리되므로 true(isnull() == True)의 개수가 나옴
```

```python
df.isnull().count()   # 해당 열의 모든 데이터 개수
```

결측값이 포함된 열 / 행 지우기

```python
df.dropna()
```

결측값이 포함된 열 지우기

```python
df.dropna(axis=1)
```

결측값을 다른 값으로 대체하기

```python
df.fillna(0)
```

앞이나 뒤의 숫자로 바꾸기

```python
df = pd.DataFrame({'a': [1, 1, 3, 4, np.nan], 'b': [2, 3, np.nan, np.nan, 4], 'c': [np.nan, 4, 7, 6, 5]})
```

```python
df
```

```python
# 뒤의 값으로 채우기
df.fillna(method='bfill')
```

```python
# 앞의 값으로 채우기
df.fillna(method='ffill')
```

```python
# limit 설정
df.fillna(method='ffill', limit=1)
```

데이터 프레임에 존재하는 결측값들을 뒤의 값으로 대체한 이후 앞의 값으로 대체하기

```python
df = pd.DataFrame({'a': [1, 1, 3, 4, np.nan], 'b': [2, 3, np.nan, np.nan, 4], 'c': [np.nan, 4, 7, 6, 5]})
df
```

```python
df.fillna(method='bfill', limit=1).fillna(method='ffill', limit=1)
```

평균으로 대체

```python
df = pd.DataFrame({'a': [1, 1, 3, 4, np.nan], 'b': [2, 3, np.nan, np.nan, 4], 'c': [np.nan, 4, 7, 6, 5]})
df
```

a column 평균으로 대체

```python
df.mean()['a']
```

```python
df.fillna(df.mean()['a'])
```

각 컬럼별 평균으로 결측값 대체

```python
df.mean()
```

```python
df.fillna(df.mean())
```

b, c의 결측값들을 각 컬럼의 데이터 평균값으로 치환

```python
df.fillna(df.mean()[['b', 'c']])
```

```python
df.mean()[['b', 'c']]
```
