# apply, map 을 활용한 데이터 변환

```python
import pandas as pd

df = pd.DataFrame({'a': [1, 2, 3, 4, 5]})
```

a가 2보다 작으면 '2 미만', 4보다 작으면 '4 미만', 4보다 크면 '4 이상'이 저장된 b 칼럼 추가

```python
df
```

```python
df['b'] = 0
df
```

```python
a = df[df['a'] < 2]
a
```

```python
df['b'][a.index] = '2 미만'
df
```

```python
b = df[(df['a'] >= 2) & (df['a'] < 4)]
b
```

```python
df['b'][b.index] = '4 미만'
df
```

```python
# 위의 warning 나타나지 않게 하기
pd.set_option('mode.chained_assignment', None)
```

```python
df['b'][b.index] = '4 미만'
df
```

```python
c = df[df['a'] >= 4]
c
```

```python
df['b'][c.index] = '4 이상'
df
```

함수 + apply 를 이용한 해결 

```python
def case_function(x):
  if x < 2:
    return '2 미만'
  elif x < 4:
    return '4 미만'
  else:
    return '4 이상'
```

```python
df['c'] = df['a'].apply(case_function)
df
```

a가 1이면 one, 2이면 two, 3이면 three, 4이면 four, 5이면 five를 출력하는 칼럼 d 생성

사용자 정의함수 사용

```python
def function(x):
  if x == 1:
    return 'one'
  elif x == 2:
    return 'two'
  elif x == 3:
    return 'three'
  elif x == 4:
    return 'four'
  elif x == 5:
    return 'five'
```

```python
df['d'] = df['a'].apply(function)
```

```python
df
```

map을 이용한 방법

```python
a = {1: 'one', 2: 'two', 3: 'three', 4: 'four', 5: 'five'}
```

```python
df['e'] = df['a'].map(a)
df
```
