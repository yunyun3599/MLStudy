# 레코드, 칼럼 추가 / 삭제

```python
import pandas as pd

df = pd.DataFrame({'a': [1, 2, 3, 4, 5], 'b': [2, 3, 2, 3, 4], 'c': [3, 4, 7, 6, 4]})
```

```python
df
```

칼럼 추가

```python
df['d'] = [1, 3, 6, 4, 8]
```

```python
df
```

```python
df['e'] = [1, 1, 1, 1, 1]
df
```

```python
df['e'] = 1   # 같은 값으로 채우고 싶을 때는 리스트를 줄 필요 없이 숫자 하나만 줘도 됨
```

```python
df
```

```python
# 칼럼끼리 계산하기 위해 컬럼의 형식 확인
df.dtypes
```

a + b - c의 결과로 이루어진 f 칼럼 추가하기

```python
df['f'] = df['a'] + df['b'] - df['c']
```

```python
df
```

칼럼 삭제

```python
df.drop(['d', 'e', 'f'], axis=1, inplace=True)    # 열 기준 삭제임을 알려주기 위해 axis = 1
```

```python
df
```

레코드 추가  
a에는 6, b에는 7, c에는 8 추가

```python
df.append({'a': 6, 'b': 7, 'c': 8})
```

```python
df = df.append({'a': 6, 'b': 7, 'c': 8}, ignore_index=True)   # append는 inplace 파라미터를 받지 않으므로 df에 다시 대입해주어야 함
df
```

a에 6, b에 8, c에 9 입력

```python
df.loc[6] = [7, 8, 9]
df
```

레코드 삭제
- 첫 번째 레코드 삭제

```python
df.drop(0, inplace=True)
df
```

```python
df.drop([1, 2], inplace=True)
df
```

첫 번째에서 네 번째 레코드를 삭제

```python
df = pd.DataFrame({'a': [1, 2, 3, 4, 5], 'b': [2, 3, 2, 3, 4], 'c': [3, 4, 7, 6, 4]})
df
```

```python
# 방법 1
df.drop([i for i in range(4)])
```

```python
# 방법 2
df.drop(df.index[:4])
```

a가 4 미만인 레코드 삭제

```python
df
```

```python
df[df['a'] < 4].index
```

```python
df.drop(df[df['a'] < 4].index)
```

a가 3 미만이고 c가 4인 레코드 삭제

```python
df[(df['a'] < 3) & (df['c'] == 4)].index
```

```python
df.drop(df[(df['a'] < 3) & (df['c'] == 4)].index)
```
