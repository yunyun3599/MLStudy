# numpy를 이용한 행렬 연산

행렬 생성

```python
import numpy as np
```

```python
a = np.array([[1, 2], [3,4]])
a
```

전치 행렬

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
a
```

```python
# 전치행렬
a.T
```

행렬 차원 확인

```python
a.shape
```

행렬 형태 변경

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.reshape(a, (3, 2))
print(b.shape)
print(b)
```

배열간 사칙연산

```python
a = np.array([[1,2,3], [4,5,6]])
b = np.array([[2,3,4], [5,6,7]])
 
print(a + b)
print(a * 3)
```

```python
print(a - b)
print(a - 2)
```

```python
print(a * b)
print(a * 2)
```

```python
print(a / b)
print(a / 2)
```

행렬의 형태가 다를 경우에는 연산 불가

```python
a = np.array([[1,2,3], [4,5,6]])
b = np.array([[2,3], [4,5]])
print(a + b)
```

행렬의 곱셈

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b))   # (1*4) + (2*5) + (3*6) = 32
```

```python
a = np.array([[1,2,3],
              [4,5,6]])
b = np.array([[1,2],
              [3,4],
              [5,6]])

print(np.dot(a, b))
```
